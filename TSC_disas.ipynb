```python
def lookup(opcode: int, funct: int):
    if (opcode == 15):
        return R_dict.get(funct, "#UD")
    else:
        return I_dict.get(opcode, "#UD")

def disas(addr: int, inst: int):
    opcode = inst >> 12
    rs = (inst >> 10) & 0b11
    rt = (inst >> 8) & 0b11
    rd = (inst >> 6) & 0b11
    funct = inst & 0b111111
    imm = (inst & 0xff) - (((inst & 0xff) >> 7) << 8)
    imm_zx = inst & 0xff
    br_target = (addr + 1 + imm) & 0xffff
    j_target = (addr & 0xf000) | (inst & 0x0fff)
    return lookup(opcode, funct).format(rs, rt, rd, imm, imm_zx, br_target, j_target)
```
